```python
# importing all the essential libraries
import pandas as pd #datamanipulation library
import matplotlib.pyplot as plt #plotting library
import numpy as np #powerful computation library
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split  
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings("ignore") #terminates the program immediately if an error occurs
```

```python
# Reading the data from the provided URL
data = pd.read_csv('DATSASET.txt')
```

```python
#examine the shape of dataset
data.shape
```

```python
#extracting the first five examples of the dataset
data.head(5)
```

```python
# Info to check for any null values and data type of Features.
data.info()
```

```python
# Describing to get quantile values 
data.describe()
```

VISUALIZING THE ENTIRE DATA

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Scored')
plt.show()
```

Model Building

```python
# Splitting the data into independent and dependent variables
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values
```

SPLITING THE DATA INTO TRAINING SET AND TEST SET

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
# Checking shapes of the splits.
print("X_train shape",X_train.shape, "\ny_train shape", y_train.shape, "\nX_test shape",X_test.shape, "\ny_test shape",y_test.shape)
```

Training the Simple Linear Regression BY TRAIN DATA

```python
reg = LinearRegression()  
reg.fit(X_train, y_train) 
```

```python
# Printing the Intercept('c') and Coefficient('m') in the regression line y = mx + c
print("Intercept : ",reg.intercept_, '\nCoefficient',reg.coef_)
```

```python
# Predicting the test set results
y_pred = reg.predict(X_test)
```

VISUALIZING THE TRAINING SET RESULTS

```python
plt.scatter(X_train, y_train, color='green')
plt.plot(X_train, reg.predict(X_train), color='red')
plt.title('Hours vs Percentage (Training set)')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.grid()
plt.show()
```

Now the model is good on training set, but we need to check the same for test set as well.

```python
# Predicting y_test from Trained Model
y_test_pred = reg.predict(X_test)
y_test_pred = pd.DataFrame(y_test_pred)
```

```python
# Now plotting a plot between predicted Test set and Actual Test
plt.scatter(X_test, y_test_pred, color='blue')
plt.plot(X_train, reg.predict(X_train), color='green')
plt.title('Hours vs Percentage (Test set)')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.grid()
plt.show()
```

THE SCORE GIVEN BY OUR MODEL FOR TRAIN SET AND TEST SET

```python
print("Train : ",reg.score(X_train,y_train)*100)
print("Test : ",reg.score(X_test,y_test)*100)
```

FINAL PREDICTION

```python
# Predicting the percentage score for a student who studies for 9.25 hours/day
hours = np.array([[9.25]])
predicted_score = reg.predict(hours)
print("Predicted Score = {}".format(predicted_score[0]))
```
